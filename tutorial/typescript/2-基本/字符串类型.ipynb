```typescript
const str = '1 day ago';
```

检查是否以指定字符串开始：

```typescript
str.startsWith('1');
```

检查是否以指定字符串结束：

```typescript
str.endsWith('ago');
```

检查是否包含指定字符串：

```typescript
str.includes('day');
```

将字符串转换为数组：

```typescript
str.split(' ');
```

获取指定索引处的字符：

```typescript
str.charAt(2);
```

获取指定开始索引到结束索引之间的字符串：

`slice(index, index)`

```typescript
'Hello, World!'.slice();
```

```typescript
'Hello, World!'.slice(6);
```

```typescript
'Hello, World!'.slice(6, 12);
```

```typescript
'Hello, World!'.slice(0, -8);
```

`substring(index, index)`

```typescript
'Hello, World!'.substring(6);
```

```typescript
'Hello, World!'.substring(6, 12);
```

`substr(index, length)`

> 已弃用

```typescript
'Hello, World!'.substr(6);
```

```typescript
'Hello, World!'.substr(6, 6);
```

返回指定字符串的索引：

```typescript
const file = 'example.ts.txt';
```

```typescript
file.indexOf('.');
```

```typescript
file.lastIndexOf('.');
```

将字符转换为大写：

```typescript
str.toUpperCase();
```

将字符转换为小写：

```typescript
str.toLowerCase();
```
