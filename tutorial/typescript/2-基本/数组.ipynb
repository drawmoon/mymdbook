在声明变量时通过 `: Type` 指定变量的类型。

字符串数组：

````typescript
const arr: string[] = ['Hello', 'World'];

console.log(arr);
````

数字数组：

````typescript
const arr: number[] = [1, 100, 0, -1];

console.log(arr);
````

联合类型数组：

````typescript
const arr: (string | number)[] = ['Hello', 123];

console.log(arr);
````

当数组中元素的类型与数组的类型不一致时，编译器则会输出错误：

```typescript
const arr: string[] = [1]; // 不能将类型“number”分配给类型“string”。
```

通过 `[v1, v2, ...]` 接收数组中的元素：

````typescript
const arr: string[] = ['Hello', 'World'];

const [hello, world] = arr;

console.log(hello, world);
````

查看指定索引处的元素：

````typescript
const fruits: string[] = ['Apple', 'Orange', 'Pear', 'Mango'];

fruits[0];
````

提取指定开始索引到结束索引之间的元素：

````typescript
fruits.slice(0, 3);
````

````typescript
fruits.slice(0, -2);
````

超出范围则返回 `undefined`：

````typescript
const a = fruits[-1];
const b = fruits[100];

console.log(a);
console.log(b);
````

移除指定索引处的元素：

````typescript
delete fruits[1];

console.log(fruits);
````

移除并返回数组的第一个元素：

````typescript
const first = fruits.shift();

console.log(first);
console.log(fruits);
````

删除并返回数组的最后一个元素：

````typescript
const last = fruits.pop();

console.log(last);
console.log(fruits);
````

将数组转换为字符串：

````typescript
const fruits: string[] = ['Apple', 'Orange'];

fruits.join(', ');
````

去除重复的数组元素：

````typescript
const same: (number | string)[] = [1, 1, 'A', 'A'];

Array.from(new Set(same));
````

## 排序

````typescript
const arr: number[] = [1, 7, 8, 12, 55];
````

配合 `Math.random()` 打乱数组中的元素：

````typescript
arr.sort(() => Math.random() - 0.5);
````

升序：

````typescript
arr.sort((a, b) => a > b ? 1 : -1);
````

降序：

````typescript
arr.sort((a, b) => a < b ? 1 : -1);
````

反转数组：

````typescript
arr.reverse();
````

## 操作多个数组

````typescript
const arr1: number[] = [1, 2, 3];
const arr2: number[] = [2, 3, 4];
````

组合：

````typescript
[...arr1, ...arr2];
````

并集：

````typescript
Array.from(new Set([...arr1, ...arr2]));
````

交集：

````typescript
arr1.filter((x) => arr1.includes(x));
````

差集：

````typescript
arr1.filter((x) => !arr2.includes(x));
````

组合数组中的数组：

````typescript
const arr: string[][] = [['a'], ['b', 'c'], ['d']];

[].concat(...arr);
````

## 计算

求和：

````typescript
const arr: number[] = [50, 20, 30];

arr.reduce((a, b) => a + b);
````

## 条件

测试所有元素是否满足条件：

````typescript
const arr: boolean[] = [true, false, true];

arr.every((x) => x);
````

测试至少一个元素是否满足条件：

````typescript
const arr: number[] = [100, 200];

arr.some((x) => x === 200);
````
